```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import seaborn as sns
```

The central limit theorem says that the sampling distribution of the mean will always be normally distributed, as long as the sample size is large enough.

To demonstrate the CLT in action, we conducted a simulation with #simulation_rounds rounds of sampling. In each round, we generated #N uniform random samples from the range [0, 1]. Next, we computed the average of each sample and normalized it. Finally, we plotted the distribution of these normalized averages and compared it with the standard normal distribution.

```python
N = 2
simulation_rounds = 1_000_000
s = np.random.uniform(low=0, high=1, size=(simulation_rounds, N)).mean(axis=1)
s = (s - 0.5) / (1 / 12 / N) ** 0.5
sns.displot(x=s, kind="kde", color='r')

x = np.arange(-4, 4, 0.01)
y = norm.pdf(x, loc=0.0, scale=1.0)
sns.lineplot(x=x, y=y)
plt.legend(["simulation", "normal dist"]);
```

In this simulation, the variable #N represents the sample size, and #simulation_rounds denotes the number of sampling rounds. By varying these parameters, we observe how the distribution of sample means evolves. As #N increases, we observe a convergence towards the standard normal distribution. This phenomenon illustrates the central limit theorem's principle: regardless of the original distribution of the data, the distribution of sample means tends towards normality with increasing sample size.
