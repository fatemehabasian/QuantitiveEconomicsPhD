```python
import numpy as np
```

```python
s = np.random.uniform(0,1,10000000)
```

In this question we assume  $x^{2} + y^{2} = 1 \Rightarrow y = f(x) = \sqrt{1-x^{2}}$

```python
def f(x): 
    return (1-x**2)**0.5
```

```python
vec_func = np.vectorize(f)
```

```python
area_under_f = np.mean(vec_func(s))
```

Area of a quarter circle with radius 1 is equal to $\pi/4$, so in order to estimate $\pi$ we have to multiple the calculated area by 4.


```python
pi = 4 * area_under_f
pi
```
